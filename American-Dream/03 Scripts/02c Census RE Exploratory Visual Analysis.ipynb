## 03 Difference in Total, Owner-occupied, and Renter-occupied housing units

To get answers for the questions related to this question, I will create a subset of the main dataset that contains the most relevent columns for this purpose. 

```python
occupancy_sub = df_re[['2012 total housing units', '2012 owner-occupied', '2012 renter-occupied', '2022 total housing units', '2022 owner-occupied', '2022 renter-occupied']]
```

```python
occupancy_sub.shape
```

<div class="alert alert-block alert-info">
    Correlation Matrix Heatmap
</div>

```python
# Using matplotlib and seaborn
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(10,10))

# Applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(occupancy_sub.corr(), annot = True, ax = ax) # The `annot` argument places the correlation coefficients onto the heatmap.
```

<div class="alert alert-block alert-info">
Scatter Plot
</div>


```python
# Create a scatterplot for 2012 total housing units and 2022 total housing units
sns.set_theme(style = 'dark')
tot_units = sns.lmplot(x = '2012 total housing units', y = '2022 total housing units', data = occupancy_sub)
plt.gca().ticklabel_format(axis = 'y', style = 'plain')
plt.gca().ticklabel_format(axis = 'x', style = 'plain')
cur_tix = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in cur_tix])
cur_x = plt.gca().get_xticks()
plt.gca().set_xticklabels(['{:,.0f}'.format(x) for x in cur_x])
plt.xticks(rotation = 45, fontsize = '12')
tot_units.fig.suptitle("Total Housing Units 2012 vs. 2022", fontsize = 12, fontdict={"weight": "bold"})
```

> There was no significant difference in total housing units between 2012 and 2022. There is a strong correlation, but no causality. 

<div class="alert alert-block alert-info">
Pair Plot
</div>


```python
# Creating a pair plot for occupancy_sub
occ_sub_pair = sns.pairplot(occupancy_sub)
occ_sub_pair.fig.suptitle("Total Housing Units 2012 vs. 2022", fontsize = 22, fontdict={"weight": "bold"}, y = 1.01)
```

> There is no significant change in the distribution of renters vs. owners, or in total units between 2012 and 2022.

## 04 Geographic Mobile Home Use

<div class = "alert alert-block alert-info">
Cat Plot
    </div>

```python
# Creating second subset for categorical plot
occ_sub_2 = df_re[['2012 mobile home', '2012 median income', '2022 mobile home', '2022 median income', 'region']]
```

```python
pd.set_option('display.max_rows', None)
occ_sub_2
```

```python
# Create a categorical plot using seaborn
g = sns.catplot(x="region", y="2012 mobile home", hue="region", data=occ_sub_2, height = 10, aspect = 1.5)
cur_tix = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in cur_tix])
g.refline(y = 7500, color = 'green')
g.fig.suptitle("Mobile Homes by Region - 2012", fontsize = 22, fontdict={"weight": "bold"}, y = 1.02)
```

```python
# Create a categorical plot using seaborn for 2022 mobile home use as a primary residence by region
g = sns.catplot(x="region", y="2022 mobile home", hue="region", data=occ_sub_2, height = 10, aspect = 1.5)
cur_tix = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.0f}'.format(x) for x in cur_tix])
g.refline(y = 7500, color = 'purple')
g.fig.suptitle("Mobile Homes by Region - 2022", fontsize = 22, fontdict={"weight": "bold"}, y = 1.02)
```

The concentration of mobile home use as a primary residence is strongest in the South region and weakest in the Northeast retion. This is an expected outcome based on weather patterns in the US. However, it appears that mobile home residence has increased in all regions over the 10 year period.

## 05 Answering Questions

### Changes from 2012 to 2022?

1. What is the percentage change in total units from 2012 to 2022?
2. Which geographic markets have seen the greatest change from 2012 to 2022?

```python
# Create a scatterplot for 2012 total housing units and 2022 total housing units
pct_diff = sns.displot(df_re['change in units %'])
cur_x = plt.gca().get_xticks()
plt.gca().set_xticklabels(['{:,.3f}'.format(x) for x in cur_x])
plt.xticks(rotation = 45, fontsize = '12')
plt.title('Percent change in total housing units 2012 - 2022')
```

```python
qq = sns.catplot(y='change in units %', x = 'region', data=df_re, height = 10)
cur_tix = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.3f}'.format(x) for x in cur_tix])
qq.refline(y = 0, color = 'red')
qq.fig.suptitle("Percent Change in Total Housing Units 2012 - 2022", fontsize = 22, fontdict={"weight": "bold"}, y = 1.02)
```

```python
zz = sns.catplot(y='change in units %', x = 'state', data=df_re, height = 5, aspect = 4)
cur_tix = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:,.3f}'.format(x) for x in cur_tix])
plt.xticks(rotation = 45, fontsize = '12')
zz.refline(y = 0, color = 'red')
zz.fig.suptitle("Percent Change in Total Housing Units 2012 - 2022", fontsize = 22, fontdict={"weight": "bold"}, y = 1.02)
```

The distribution of all values appears to be in a normal distribution shape centered on zero. There are few extreme outliers, which indicate large increases or decreases in individual counties. There are increases and decreases in many counties, but it appears that the decreases in some counties are offset by increases in other counties. 

This distribution is further reflected at the regional and state levels. 

On the whole, total housing remained flat between 2012 and 2022.
